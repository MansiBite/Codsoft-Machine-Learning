```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas dataframe
credit_data = pd.read_csv('/content/creditcard.csv')
```

```python
# to check number of missing value in a dataset
credit_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudlent transaction
credit_data['Class'].value_counts()
```

```python
#This dataset is highly unblanced
#0 is normal transaction
#1 is fraudlent transaction
# separating the data for analysis
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
credit_data.groupby('Class').mean()
```

```python
#Build sample dataset conatining similar distribution of normal transaction and fraudlent transaction
#number of normal transaction is 19397
#number of fraudlent transation is 68
legit_sample=legit.sample(n=68)
```

```python
#concatenating two dataframe
new_dataframe=pd.concat([legit_sample,fraud],axis=0) #rows
```

```python
new_dataframe.head()
```

```python
new_dataframe.tail()
```

```python
new_dataframe['Class'].value_counts()
```

```python
new_dataframe.groupby('Class').mean()
```

```python
#splitting the data into features and targets
x=new_dataframe.drop(columns='Class',axis=1)
y=new_dataframe['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#split data into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#Model Training
#logistic Regression
model = LogisticRegression()
```

```python
#training the logistic Regression Model with Training Data
model.fit(x_train, y_train)
```

```python
#Model Evaluation
#Accuracy Score
#accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy Score will be on training data :",training_data_accuracy)
```

```python
#Accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy score on this testing data :",test_data_accuracy)
```
